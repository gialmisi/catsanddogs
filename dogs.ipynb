```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from parse import read_excel
from desdeo_tools.utilities import fast_non_dominated_sort_indices
import plotly.express as px
```

```python
df = read_excel(fname="Canine_dataset.xlsx").dropna()
```

```python
objectives = ["insecurity_score", "training_focus_score", "activity_playfulness_score", "aggressiveness_dominance_score", "human_sociability_score", "dog_sociability_score", "perseverance_score"]
```

```python

directions = [1, -1, -1, 1, -1, -1, -1]
to_max = ["training_focus_score", "activity_playfulness_score", "human_sociability_score", "dog_sociability_score", "perseverance_score"]

df[to_max] = -df[to_max]
data = df[objectives].to_numpy()

nadir, ideal = np.max(data, axis=0), np.min(data, axis=0)

df
```

```python
fronts = fast_non_dominated_sort_indices(data)
pf_data = df.iloc[fronts[0]]
```

```python
from collections import Counter

dogs = dict(Counter(pf_data["breed_group"]))

df[objectives]
dogs
```

```python
pf_data = pf_data[["breed_group"] + objectives]
pf_data
```

```python
# normalized_df=(df-df.min())/(df.max()-df.min())

pf_data[objectives] = ((pf_data[objectives] - pf_data[objectives].min()) / (pf_data[objectives].max() - pf_data[objectives].min()) * 100).round(decimals=2)
pf_data[to_max] = -pf_data[to_max]
```

```python
pf_data.round(decimals=2)
```

```python
pf_data.to_csv("./data/pf_dog_data.csv", index=False)
```
